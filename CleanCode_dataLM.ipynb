# DATASET MERGE

```python

# demo + pd_de + cop_u +  pd_ca
cust = []  #고객 유형 만드는 함수
for i in range(len(demo)):
    if demo.성별[i] == '여성':
        if demo.나이대[i] == '10대':
            cust.append('F10')
        elif demo.나이대[i] == '20대':
            cust.append('F20')
        elif demo.나이대[i] == '30대':
            cust.append('F30')
        elif demo.나이대[i] == '40대':
            cust.append('F40')
        elif demo.나이대[i] == '50대':
            cust.append('F50')
        elif demo.나이대[i] == '60대':
            cust.append('F60')
        else:
            cust.append('F70')
    else:
        if demo.나이대[i] == '10대':
            cust.append('M10')
        elif demo.나이대[i] == '20대':
            cust.append('M20')
        elif demo.나이대[i] == '30대':
            cust.append('M30')
        elif demo.나이대[i] == '40대':
            cust.append('M40')
        elif demo.나이대[i] == '50대':
            cust.append('M50')
        elif demo.나이대[i] == '60대':
            cust.append('M60')
        else:
            cust.append('M70')
demo['성별나이'] = cust
```

```python
# 요일 추출.
data1234['거래날짜'] = pd.to_datetime(data1234['거래날짜'],format='%Y%m%d')
data1234['거래요일'] = data1234['거래날짜'].dt.dayofweek #0:월요일
data1234['거래요일'].replace({0:'월',1:'화',2:'수',3:'목',4:'금',5:'토',6:'일'})
```

```python
# demo + pd_de + cop_u + pd_ca + lpay
lpay_cu = lpay['고객번호'].drop_duplicates().reset_index(drop=True)
lpay_cu = lpay_cu.values.tolist()

data1234_cu = pd.DataFrame(data1234['고객번호'].drop_duplicates().reset_index(drop=True), columns=["고객번호"])
lpay_list = data1234_cu

l = []
for i in range(len(lpay_cu)):
    l.append(lpay_cu[i])


l_cust = []
for i in tq(range(len(data1234_cu))):
    if data1234_cu.고객번호[i] in l:
        l_cust.append('1')
    else:
        l_cust.append('0')
lpay_list = pd.DataFrame(lpay_list, columns=['고객번호', '회원유무'])
lpay_list['회원유무'] = l_cust
print(lpay_list)

data12346 = pd.merge(data1234, lpay_list)
```
